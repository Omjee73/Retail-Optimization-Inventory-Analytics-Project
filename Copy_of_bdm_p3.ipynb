```python
### ZOOMED VERSION of Red dots...

import pandas as pd
import matplotlib.pyplot as plt

# Given Data
data = {
    "Product": [
        "Ayur Herbals Cream", "Bikaji Bhujia 200g", "Bikaji Bhujia 400g", "Bikano Bikaneri Bhujia",
        "Bournvita", "Britannia Cake", "Britannia Cookies", "Brooke Bond Red Label Tea",
        "Cadbury Celebrations", "Ching's Masala", "Chocolate", "Closeup Toothpaste", "Cookies",
        "Dabur Amla Hair Oil", "Dabur Red Toothpaste", "Dettol Handwash", "Dettol Soap",
        "Diaper Pants", "Dove Moisture", "Dove Shampoo", "Dove soap", "Dr Oetker Veg Mayonnise",
        "Dukes (snacks)", "Eveready cell", "Face Wash", "Ghadi Detergent Powder", "Gillette Shaving Foam",
        "Glow & Lovely Facewash", "Glucon D", "Himalaya Facewash", "Horlicks", "Indulekha Oil",
        "Indulekha Shampoo", "Kellogg's", "Kissan Tomato Ketchup", "Lipton Green Tea", "Lotte Candy",
        "Mama Earth Face Wash", "Oreo Biscuits", "Parachute Coconut Oil", "Parle Candy",
        "Patanjali Facewash", "Patanjali Shampoo", "Patanjali Toothpaste", "Priyagold Biscuits",
        "Saffola Oats", "Sunsilk Shampoo", "Surf Excel Detergent", "Taj Mahal Tea", "Tata Tea",
        "Vaseline", "Vim Dishwash Bar"
    ],
    "Sell_Quantity": [
        14, 10, 19, 14, 14, 14, 19, 9, 66, 46, 3, 26, 141, 13, 8, 6, 24, 4, 3, 7, 10, 15, 17, 2, 6,
        7, 2, 20, 15, 21, 9, 3, 8, 29, 25, 12, 3, 11, 50, 74, 39, 7, 17, 29, 20, 24, 1, 47, 6, 20, 14, 11
    ],
    "Purchase_Quantity": [
        14, 10, 24, 14, 24, 20, 136, 9, 76, 52, 3, 36, 174, 13, 9, 6, 24, 4, 3, 7, 15, 16, 17, 2, 6,
        10, 2, 32, 26, 21, 9, 3, 9, 34, 43, 12, 3, 11, 58, 92, 49, 7, 17, 49, 20, 25, 1, 74, 6, 21, 15, 11
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Step 1: Identify Out of Stock Products
out_of_stock = df[df['Sell_Quantity'] == df['Purchase_Quantity']]

# Visualize Stock-Out Frequency vs Demand
plt.figure(figsize=(12, 8))
plt.scatter(out_of_stock['Sell_Quantity'], out_of_stock['Purchase_Quantity'], color='red', label='Stock-Out Products')

# Adjust Label Positions Dynamically
label_positions = {}

for i in range(len(out_of_stock)):
    x = out_of_stock['Sell_Quantity'].iloc[i]
    y = out_of_stock['Purchase_Quantity'].iloc[i]
    product = out_of_stock['Product'].iloc[i]

    # If this point is already occupied, adjust position dynamically
    if (x, y) in label_positions:
        if label_positions[(x, y)] == "left":
            plt.text(x - 0.5, y - 0.5, product, fontsize=8, ha='right', va='center')  # Move left-down
            label_positions[(x, y)] = "left-lower"
        elif label_positions[(x, y)] == "left-lower":
            plt.text(x + 0.5, y, product, fontsize=8, ha='left', va='center')  # Move right
            label_positions[(x, y)] = "right"
        elif label_positions[(x, y)] == "right":
            plt.text(x + 0.5, y - 0.5, product, fontsize=7, ha='left', va='center')  # Move right-down (reduce font)
            label_positions[(x, y)] = "right-lower"
        else:
            plt.text(x, y, product, fontsize=7, ha='center', va='center')  # As last resort, place in center
    else:
        plt.text(x - 0.5, y, product, fontsize=8, ha='right', va='center')  # Default placement on left
        label_positions[(x, y)] = "left"

# Customize Plot
plt.title('Demand vs Stock-Out (Zoomed version of red dots)', fontsize=14)
plt.xlabel('Sell Quantity (Demand)', fontsize=12)
plt.ylabel('Purchase Quantity (Stock-Out Level)', fontsize=12)
plt.grid(True, alpha=0.5)
plt.tight_layout()

# Show the Graph
plt.show()

```
